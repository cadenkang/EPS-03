```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
RCP85 = pd.read_table('RCP85 (1).txt', header=None)
# sep = ‘\s+’
RCP85 = RCP85[0].str.split(expand=True)
RCP85 = RCP85.rename({0:'Year', 1:'CO2 Annual Emissions', 2:'Atmospheric CO2', 3:'Total alkalinity (µmol/kg)'}, axis=1)
RCP85
```

```python
RCP85['Year'] = RCP85['Year'].astype(int)
RCP85['Year'] *= -1
RCP85
```

```python
RCP85['CO2 Annual Emissions'] = RCP85['CO2 Annual Emissions'].astype(float)
RCP85['Atmospheric CO2'] = RCP85['Atmospheric CO2'].astype(float)
RCP85['Total alkalinity (µmol/kg)'] = RCP85['Total alkalinity (µmol/kg)'].astype(float)
RCP85
```

```python
RCP85 = RCP85.set_index('Year')
RCP85
```

```python
fig = plt.figure()
axl = fig.add_subplot(111)
annual = RCP85.iloc[:,0]
atmo = RCP85.iloc[:,1]
alka = RCP85.iloc[:,2]
data = (annual, atmo, alka)

for i in data:
    lines = axl.plot(i, label=i.name)

axl.legend(title='Legend')
axl.grid(color='silver', b=True, which='both')
axl.set_facecolor('aliceblue')
axl.set_title('RCP 8.5', fontsize=15, fontweight='bold')
axl.set(xlabel='Year')
```
