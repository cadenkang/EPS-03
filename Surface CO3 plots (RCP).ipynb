```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
RCP26 = pd.read_table('RCP26.txt', header=None, sep = '\s+')
RCP26[0] = RCP26[0].astype(int)
RCP26[0] *= -1
RCP26[4] = RCP26[4].astype(float)
for n in range(1,6):
    if n != 4:
        RCP26 = RCP26.drop(n,axis=1)
RCP26 = RCP26.rename({0:'Year', 4:'RCP 2.6'}, axis=1)
RCP26 = RCP26.set_index('Year')
RCP26
```

```python
RCP45 = pd.read_table('RCP45.txt', header=None, sep = '\s+')
RCP45[0] = RCP45[0].astype(int)
RCP45[0] *= -1
RCP45[4] = RCP45[4].astype(float)
for n in range(1,6):
    if n != 4:
        RCP45 = RCP45.drop(n,axis=1)
RCP45 = RCP45.rename({0:'Year', 4:'RCP 4.5'}, axis=1)
RCP45 = RCP45.set_index('Year')
RCP45
```

```python
RCP6 = pd.read_table('RCP6.txt', header=None, sep = '\s+')
RCP6[0] = RCP6[0].astype(int)
RCP6[0] *= -1
RCP6[4] = RCP6[4].astype(float)
for n in range(1,6):
    if n != 4:
        RCP6 = RCP6.drop(n,axis=1)
RCP6 = RCP6.rename({0:'Year', 4:'RCP 6.0'}, axis=1)
RCP6 = RCP6.set_index('Year')
RCP6
```

```python
RCP85 = pd.read_table('RCP85 (2).txt', header=None, sep = '\s+')
RCP85[0] = RCP85[0].astype(int)
RCP85[0] *= -1
RCP85[4] = RCP85[4].astype(float)
for n in range(1,6):
    if n != 4:
        RCP85 = RCP85.drop(n,axis=1)
RCP85 = RCP85.rename({0:'Year', 4:'RCP 8.5'}, axis=1)
RCP85 = RCP85.set_index('Year')
RCP85
```

```python
fig = plt.figure()
axl = fig.add_subplot(111)
new26 = RCP26.iloc[40:,0]
new45 = RCP45.iloc[40:,0]
new6 = RCP6.iloc[40:,0]
new85 = RCP85.iloc[40:,0]
data = (new26, new45, new6, new85)

for i in data:
    lines = axl.plot(i, label=i.name, linewidth=3.5)

axl.legend(title='RCP Scenarios', prop={'size': 10}, title_fontsize=12)
axl.grid(color='silver', b=True, which='both')
axl.set_facecolor('aliceblue')
axl.set_title('Surface $CO_{3}$$^{2-}$ Concentration', fontsize=18, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.xticks(fontsize=11)
plt.ylabel('$CO_{3}$$^{2-}$ (µmol/kg)', fontsize=14)
plt.yticks(fontsize=11)

params = {'mathtext.default':'regular'}          
plt.rcParams.update(params)
```
