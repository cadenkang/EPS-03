```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# control RCP 2.6
control26 = pd.read_table('RCP_26.txt', header=None, sep='\s+')
control26[0] = control26[0].astype(int)
control26[0] *= -1
for n in range(1, 9):
    control26[n] = control26[n].astype(float)
control26[1] *= 1000000000
control26 = control26.rename({0:'Year', 1:'CO2 Emission', 2:'Atmospheric CO2', 3:'Surface Alkalinity', 
                              4:'Surface Carbonate Concentration', 5:'Omega at Surface', 6:'Surface pH',
                              7:'Annual Alkalinity Added', 8:'Total Alkalinity Added'}, axis=1)
control26 = control26.set_index('Year')
control26
```
