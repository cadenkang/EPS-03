```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# IAM models shows the baseline down found in the carbon brief article
IAMmodels = pd.read_csv('SSP_IAM_V2_201811.csv')
CMIP6 = pd.read_csv('SSP_CMIP6_201811.csv')
```

```python
world_IAM = IAMmodels[IAMmodels.REGION=='World']
world_CMIP6 = CMIP6[CMIP6.REGION=='World']
```

```python
CO2_IAM = world_IAM[world_IAM.VARIABLE=='Emissions|CO2']
temp_IAM = world_IAM[world_IAM.VARIABLE=='Diagnostics|MAGICC6|Temperature|Global Mean']
CO2_CMIP6 = world_CMIP6[world_CMIP6.VARIABLE=='CMIP6 Emissions|CO2']
```

```python
# probably would like to index by Year and only baseline scenariostemp_IAM
CO2_IAM.drop('REGION',axis=1)
```

```python
CO2_IAM
```

```python
# taking just the baseline
baselineCO2 = CO2_IAM[CO2_IAM['SCENARIO'].str.contains('Baseline')]
# now I want just the marker models (in order IMAGE,MESSAGE,AIM,GCAM,REMIND)
# remove Witch
baselineCO2_clean = baselineCO2[baselineCO2['MODEL']!='WITCH-GLOBIOM']
# drop useless columns
baselineCO2_clean = baselineCO2_clean.drop('REGION',axis=1)
baselineCO2_clean = baselineCO2_clean.drop('VARIABLE',axis=1)
# try to remove SSP 2-5 for MESSAGE
```

```python
baselineCO2_clean
```

```python
baselineCO2_clean.set_index('SCENARIO')
```

```python
test = baselineCO2_clean.drop(baselineCO2_clean[(baselineCO2_clean['MODEL'] == 'AIM/CGE') & (baselineCO2_clean['SCENARIO'] == 'SSP1-Baseline')].index)
```

```python
#baselineCO2_clean.drop(baselineCO2_clean[(baselineCO2_clean['MODEL'] == 'REMIND-MAGPIE') & (baselineCO2_clean['SCENARIO'] == 'SSP2-Baseline')].index)
#df_new = df.drop(df[(df['col_1'] == 1.0) & (df['col_2'] == 0.0)].index)
# example cleaning for IMAGE SSP1
for n in range(2,6):
    test = test.drop(test[(test['MODEL'] == 'IMAGE') & (test['SCENARIO'] == 'SSP{}-Baseline'.format(n))].index)
    print('Removing','SSP{}-Baseline'.format(n))
```

```python
# cleaning for MESSAGE-GLOBIOM SSP2
for n in range(1,6):
    if n != 2:
        test = test.drop(test[(test['MODEL'] == 'MESSAGE-GLOBIOM') & (test['SCENARIO'] == 'SSP{}-Baseline'.format(n))].index)
        print('Removing','SSP{}-Baseline'.format(n))
```

```python
# cleaning for AIM SSP3
for n in range(1,6):
    if n != 3:
        test = test.drop(test[(test['MODEL'] == 'AIM/CGE') & (test['SCENARIO'] == 'SSP{}-Baseline'.format(n))].index)
        print('Removing','SSP{}-Baseline'.format(n))
```

```python
# cleaning for GCAM SSP4
for n in range(1,6):
    if n != 4:
        test = test.drop(test[(test['MODEL'] == 'GCAM4') & (test['SCENARIO'] == 'SSP{}-Baseline'.format(n))].index)
        print('Removing','SSP{}-Baseline'.format(n))
```

```python
# cleaning for REMIND SSP5
for n in range(1,5):
    test = test.drop(test[(test['MODEL'] == 'REMIND-MAGPIE') & (test['SCENARIO'] == 'SSP{}-Baseline'.format(n))].index)
    print('Removing','SSP{}-Baseline'.format(n))
```

```python
# still need to clean the rest of the models but eventually we would probably want years for the rows 
test.transpose()
```

```python
sspData = test.rename({8973:'SSP3', 26685:'SSP4', 34567:'SSP1', 55157:'SSP2', 70052:'SSP5'})
sspData
```

```python
sspData = sspData.drop('MODEL',axis=1)
sspData = sspData.drop('SCENARIO',axis=1)
sspData = sspData.drop('UNIT',axis=1)
sspData
```

```python
sspData.transpose()
```

```python
fig = plt.figure()
axl = fig.add_subplot(111)
SSP3 = sspData.iloc[0]
SSP4 = sspData.iloc[1]
SSP1 = sspData.iloc[2]
SSP2 = sspData.iloc[3]
SSP5 = sspData.iloc[4]
SSPS = (SSP1, SSP2, SSP3, SSP4, SSP5)

for i in SSPS:
    lines = axl.plot(i, label=i.name)

axl.legend(title='SSPs')
axl.grid(color='silver', b=True, which='both')
axl.set_facecolor('paleturquoise')
axl.set_title('CO2 emission for each SSP baseline scenario', fontsize=13, fontweight="bold")
axl.set(xlabel='Year', ylabel='CO2 emission (Mt CO2/yr)')
```
